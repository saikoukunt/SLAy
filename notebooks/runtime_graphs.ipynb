```python
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
record_length = [41, 53, 66, 82]
pass_clust = [304, 258, 1811, 633]


cache_spikes_time = [17, 55, 538, 136]
cross_sim_time = [134, 129, 1272, 548]
xcorr_time = [9, 34, 517, 152]
ref_pen_time = [12, 70, 647, 197]
merge_time = [14, 44, 247, 50]
total_time = [190, 343, 3264, 1067]
```

```python
plt.figure(figsize=(8,4))
plt.plot(record_length, cache_spikes_time, label="cache spikes", marker='o');
plt.plot(record_length, cross_sim_time, label="cross similarity", marker='o');
plt.plot(record_length, xcorr_time, label="cross correlation");
plt.plot(record_length, ref_pen_time, label="refractory period penalty", marker='o');
plt.plot(record_length, merge_time, label="merging", marker='o');
plt.grid()
plt.legend()
plt.xlabel("Recording duration (mins)"); plt.ylabel("seconds");
plt.title("Stage runtimes vs. recording duration");
plt.savefig("duration.png")
```
